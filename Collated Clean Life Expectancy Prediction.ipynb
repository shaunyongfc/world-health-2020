Keras model to predict life expectancy from given healthy lifestyle data after proper data cleaning.

```python
from my_utils import my_utils
```

```python
df = my_utils.df_latest_cleaned()
```

```python
import numpy as np
import pandas as pd
from keras.models import Model
from keras.layers.normalization import BatchNormalization
from keras.layers import Input, Dense, Dropout,  Activation
from keras.optimizers import Adam

INIT_LR = 1e-3
BS = 5
EPOCHS = 1000

def create_model(input_shape):
    """
    Create the neural network model.
    """
    inputs = Input(shape=input_shape)
    # First fully connected layer
    x = Dense(100)(inputs)
    x = Activation("relu")(x)
    x = BatchNormalization()(x)
    x = Dropout(0.5)(x)
    # Second fully connected layer
    x = Dense(50)(x)
    x = Activation("relu")(x)
    x = BatchNormalization()(x)
    x = Dropout(0.5)(x)
    # Final fully connected layer
    x = Dense(1)(x)
    model = Model(inputs=inputs, outputs=x)
    opt = Adam(lr=INIT_LR, decay=INIT_LR / EPOCHS)
    model.compile(optimizer=opt,loss='mean_squared_error',metrics = ['mean_squared_error'])
    return model
model = create_model((35,))
```

```python
X = df.drop(['Life expectancy at birth (years)'], axis=1).to_numpy()
y = df['Life expectancy at birth (years)'].to_numpy()
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
```

```python
h = model.fit(X_train, y_train, validation_data=(X_test, y_test), epochs=EPOCHS, batch_size=BS)
```

```python
location_predict = 'Singapore'
X_predict = df.loc[location_predict].drop('Life expectancy at birth (years)').to_numpy()
y_predict = df.loc[location_predict, 'Life expectancy at birth (years)']
```

```python
model.predict(np.expand_dims(X_predict, 0))
```

```python
def predict_location(location_predict):
    X_predict = df.loc[location_predict].drop('Life expectancy at birth (years)').to_numpy()
    y_predict = df.loc[location_predict, 'Life expectancy at birth (years)']
    print(f"Prediction: {model.predict(np.expand_dims(X_predict, 0))[0][0]: .2f}\nCorrect Output: {y_predict}")
```
